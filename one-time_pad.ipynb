```python
import os
```

```python
plaintext = b'foobar'
```

```python
key = os.urandom(len(plaintext))
print(key)
```

```python
# This hint to use zip is elegant, see https://stackoverflow.com/a/29409299.
ciphertext = bytes(p ^ k for p, k in zip(plaintext, key))
print(ciphertext)
```

```python
print(bytes(c ^ k for c, k in zip(ciphertext, key)))
```
